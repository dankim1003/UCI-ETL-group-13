```python
import pandas as pd
from sqlalchemy import create_engine 
```

```python
faostat_file = "../Resources/FAOSTAT_data.csv"
faostat_df = pd.read_csv(faostat_file) 
faostat_df.head()
```

```python
#create a altered dataframe from specific columns
faostat_cols = ["Country", "Year", "Element", "Item", "Year", "Unit", "Value"]
faostat_transformed = faostat_df[faostat_cols].copy()

# Cleaning data of possible duplicates
faostat_transformed.set_index("Country", inplace=True)

faostat_transformed.head()
```

```python
data_file = "../Resources/data.csv"
data_df = pd.read_csv(data_file) 
data_df.head()
```
