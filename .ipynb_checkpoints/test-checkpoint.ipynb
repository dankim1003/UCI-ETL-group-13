```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
fao_data = "Resources/faostat.csv"
obesity_data = "Resources/obesity.csv"
```

```python
fao_df = pd.read_csv(fao_data)
obesity_df = pd.read_csv(obesity_data)
```

```python
fao_sort = fao_df[["Country", "Item", "Year", "Value" ]]
```

```python
obesity_sort = obesity_df[["Country", "Year", "Obesity (%)", "Sex" ]]
```

```python
merged_data = pd.merge(fao_sort, obesity_sort, on="Country", how="left")
merged_data
```

```python
new_merged_data =  merged_data[['Country', 'Item', 'Year_x', 'Value', 'Obesity (%)', 'Sex']]
new_merged_data.head()
```
